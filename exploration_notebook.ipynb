```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

mitbih_test_df = pd.read_csv("mitbih_test.csv")
mitbih_train_df = pd.read_csv("mitbih_train.csv")
ptbdb_abnormal_df = pd.read_csv("ptbdb_abnormal.csv")
ptbdb_normal_df = pd.read_csv("ptbdb_normal.csv")
```

```python
new_column_names = range(1, len(mitbih_train_df.columns) + 1)
mitbih_train_df.columns = new_column_names
mitbih_train_df.describe()
# mitbih_train_df[188].value_counts()
```

```python
new_column_names = range(1, len(mitbih_test_df.columns) + 1)
mitbih_test_df.columns = new_column_names
mitbih_test_df.describe()
# mitbih_train_df[188].value_counts()
```

```python
new_column_names_2 = range(1, len(mitbih_test_df.columns) + 1)
mitbih_test_df.columns = new_column_names_2
# print(mitbih_test_df.head())
print(mitbih_test_df[188].value_counts())
```

```python
# Grafic representation of the Categories - MIT-BIH - Train
# Group by the category in the column named '188'
grouped = mitbih_train_df.groupby(188)
# Calculate the mean for each category across the 187 columns
means = grouped.mean()
# Plot the means for each category
plt.figure(figsize=(12, 8))
for category in means.index:
    plt.plot(range(1, 188), means.loc[category], label=f'Category {category}')
plt.xlabel('Data Points')
plt.ylabel('Mean Values')
plt.title('Mean EKG Data by Category - MIT-BIH_train')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Grafic representation of the Categories - MIT-BIH- Test
# Group by the category in the column named '188'
grouped = mitbih_test_df.groupby(188)
# Calculate the mean for each category across the 187 columns
means = grouped.mean()
# Plot the means for each category
plt.figure(figsize=(10, 6))
for category in means.index:
    plt.plot(range(1, 188), means.loc[category], label=f'Category {category}')
plt.xlabel('Data Points')
plt.ylabel('Mean Values')
plt.title('Mean EKG Data by Category - MIT-BIH_test')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PTBDB Normal
new_column_names = range(1, len(ptbdb_normal_df.columns) + 1)
ptbdb_normal_df.columns = new_column_names
#ptbdb_normal_df.head()
ptbdb_normal_df[188].describe()
```

```python
# PTBDB Abnormal
new_column_names = range(1, len(ptbdb_abnormal_df.columns) + 1)
ptbdb_abnormal_df.columns = new_column_names
#ptbdb_abnormal_df.head()
ptbdb_abnormal_df[188].describe()
#ptbdb_abnormal_df.isna().sum().sum()
```

```python
# Grafic representation of the Categories - ptbdb_normal
# Group by the category in the column named '188'
grouped = ptbdb_normal_df.groupby(188)
# Calculate the mean for each category across the 187 columns
means = grouped.mean()
# Plot the means for each category
plt.figure(figsize=(12, 8))
for category in means.index:
    plt.plot(range(1, 188), means.loc[category], label=f'Category {category}')
plt.xlabel('Data Points')
plt.ylabel('Mean Values')
plt.title('Mean EKG Data by Category - ptbdb_normal')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# PTBDB Verteilung Spalte 1
plt.figure(figsize=(10, 6))

plt.subplot(1, 2, 1)
plt.hist(ptbdb_abnormal_df.iloc[:, 0], bins=50, alpha=0.7, label='Abnormal')
plt.title('Distribution of First Column - PBTBDB_Abnormal')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(ptbdb_normal_df.iloc[:, 0], bins=50, alpha=0.7, label='Normal')
plt.title('Distribution of First Column - PTBDB_Normal')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
